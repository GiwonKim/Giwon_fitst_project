```python
class H():

  def __init__(self, w):
    self.w = w

  def forward(self, x):
    return self.w * x

# class 개념 오래되서 기억 안나는 부분 복습해놓기     

def cost(h,x,y):
  error= 0
  for i in range(len(X)):
    error +=(h.forward(X[i])-Y[i])**2
    #print(error)
  error=error/len(X)
  return error

h=H(4)
# H(x)=4x 라는 가설을 설정 

cost(h,X,Y)


```

Hypothesis 생성, cost functino 생성 , w가 4 인경우에 대한 error = 222.2

```python
list_w=[]
list_c=[]

for i in range(-100, 100):
  w=i*0.1 
  h=H(w)
  c=cost(h,X,Y)
  list_w.append(w)
  list_c.append(c)

plt.figure(figsize=(2,2))
plt.xlabel("w")
plt.ylabel("cost")
plt.plot(list_w,list_c)
```

w를 0.1~10까지 넣고 각각의 hypothesis에 따른 cost function의 값을 plot 함


```python
def cal_grad(w,cost):
  h=H(w)
  cost1 = cost(h,X,Y)
  eps = 0.001
  h=H(w+eps)
  cost2=cost(h,X,Y)
  dcost = cost2-cost1
  dw=eps
  grad=dcost/dw
  return grad


w=4
lr=0.01
# w외 초기값, learning rate 의 초기값

for i in range(100):
  w -= lr*(cal_grad(w,cost))
  print(w)
# gradient descent 의 메인 식 


```
